```python
# Bibliothèques de base
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def initialisation(X):
    W = np.random.randn(X.shape[1], 1)
    b = np.random.randn(1)
    return (W, b)
```

```python
def model (X, W, b):
    Z = X.dot(W) + b
    A = 1/(1+ np.epx(-Z))  # fonction d'activation
    return A
```

```python
def cost (A, y):
    return 1 / len(y) * np.sum(-y * np.log(A) - (1-y) * np.log(1 - A))
```

```python
def gradients(A, X, y):
    dW = 1 / len(y) * np.dot(X.T, A - y)
    db = 1 / len(y) * np.sum(A - y)
    return(dW, db)
```

```python
def update_param(dW, db, W, b, learning_rate):
    W = W - learning_rate * dW
    b = b - learning_rate * db
    return (W, b)
```

## Perceptron 

```python
def artificial_neuron (X, y, learning_rate = 0.1, n_iter = 100):
    # initialisation de W et b
    W, b = initialisation(X)
    Loss = []
    
    # Entrainement du modèle
    for i in range(n_iter):
        A = model(X, W, b)
        Loss.append(cost(A, y))
        dW, db = gradients( A, X, y)
        W, b = update_param(dW, db, W, b, learning_rate)
    
    plt.plot(Loss) # Pour visualiser si l'apprentissage se déroule convenablement
    
    return (W, b)
```

### La décision du neurone

```python
def predict(X, W, b, seuil = 0.5):
    A = model(X, W, b)
    return A >= seuil
```
